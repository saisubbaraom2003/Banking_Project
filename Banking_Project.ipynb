#Data Preprocessing

##Importing Required Libralies

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

##Encoding

```python
df=df.drop(['id'],axis=1)#Droping Id which is useless
```

```python
label_encoder=LabelEncoder()
df['Group']=label_encoder.fit_transform(df['Group'])
```

```python
train=df[df["type"]=='train']
test=df[df["type"]=='test']
```

```python
train=train.drop(["type"],axis=1)
test=test.drop(["type"],axis=1)
test=test.drop(["Target"],axis=1)
```

##Scaling

```python
scaler=MinMaxScaler()
df_scaled=pd.DataFrame(scaler.fit_transform(train),columns=train.columns)
```

```python
df_scaled
```

```python
x=df_scaled.drop(['Target'],axis=1)
y=df_scaled['Target']
```

##Spliting Data into Training and Testing

```python
df_scaled["Target"].value_counts()
```

Since the target class is imbalence so we used stratification for balancing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

#Building Maching Learning (Traditional Algorithm)

##Importing Required Libralies

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
# evaluation matrics
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    XGBClassifier(),
    KNeighborsClassifier(),
    SVC(),
    GaussianNB(),
    VotingClassifier(
        estimators=[
            ('logit', LogisticRegression()),
            ('dtree', DecisionTreeClassifier()),
            ('rforest', RandomForestClassifier()),
            ('xgboost', XGBClassifier()),
            ('knn', KNeighborsClassifier()),
            ('svm', SVC()),
            ('nbt', GaussianNB())
        ],
        voting='hard'
    )
]
```

```python
scores = {}
for model in models:
    model.fit(x_train, y_train)
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    scores[type(model).__name__] = {"Train Accuracy": train_accuracy, "Test Accuracy": test_accuracy}
```

```python
accuracy_table = pd.DataFrame.from_dict(scores, orient="index").reset_index()
accuracy_table.rename(columns={"index": "Model"}, inplace=True)
```

```python
accuracy_table
```

Every Model is giving almost same Accuracy So i am going with one of them

```python
model=RandomForestClassifier()
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
print("Training Score",round(accuracy_score(y_train, y_train_pred),2))
print("Testing Score",accuracy_score(y_test, y_test_pred))
```

##Cross Validation

```python
cvs=cross_val_score(model,x_train,y_train,cv=10)
print("Training Mean",round(cvs.mean(),2))
```
